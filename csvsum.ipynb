```python
seoul_car = pd.read_csv('/content/drive/MyDrive/ai+x 기초/car.csv', sep = ',', encoding = 'UTF-8')
seoul_car
```

```python
seoul_subway = pd.read_csv('/content/drive/MyDrive/ai+x 기초/subway.csv', sep = ',', encoding = 'cp949')
seoul_subway.head()
```

```python
a = pd.merge(seoul_sum,seoul_car,how='inner', on='자치구')
b = pd.merge(a,seoul_subway,how = 'inner', on = '자치구')
seoul_all = pd.merge(b,seoul_area, how='inner', on = '자치구')
seoul_all = seoul_all[['자치구','인구','정류장수','노선수','승차총승객수','하차총승객수','65세이상고령자','역수','등록 자동차 수','면적','구성비']]
seoul_all.sort_values(by = '65세이상고령자', ascending = False)
```
